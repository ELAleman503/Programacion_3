```python
import pandas as pd
import matplotlib.pyplot as meGraficas

tarjetasCredito = pd.read_csv("/content/trx_tarjetas_credito.csv", sep=",")
```

```python
getGender = pd.crosstab(index=tarjetasCredito["Gender"],columns="Gender")
print(getGender)

pieGender = getGender.loc[getGender.index]["Gender"]
meGraficas.pie(pieGender, autopct='%1.1f%%', shadow=True, startangle=90,)
meGraficas.title("Uso de tarjetas por hombre y mujeres")
meGraficas.show
getGender.to_excel("Credit_Card_For_Gender.xlsx")
```

```python
getCityType = tarjetasCredito.groupby(["City", "Card Type"]).sum()
print(getCityType)

pieCityType = getCityType.loc[getCityType[:10].index]
meGraficas.pie(pieCityType, autopct='%1.1f%%', shadow=True, startangle=90,)
meGraficas.title("Tipos de tarjetas de credito por ciudades")
meGraficas.show

getCityType.to_excel("Card_Type_For_City.xlsx")
```

```python
getUse = pd.crosstab(index=tarjetasCredito["Exp Type"],columns="Use")
print(getUse)

pieUse = getUse.loc[getUse.index]["Use"]
meGraficas.pie(pieUse, autopct='%1.1f%%', shadow=True, startangle=90,)
meGraficas.title("Uso de tarjetas por gasto")
meGraficas.show
getUse.to_excel("Credit_Card_For_Use.xlsx")
```
